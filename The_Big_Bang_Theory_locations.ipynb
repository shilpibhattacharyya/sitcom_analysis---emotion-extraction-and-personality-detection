```python
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
df=pd.read_csv('big_bang_theory_dataset.csv')
```

```python
df[df['Location']=='The same']
```

```python
#print(scene_text)
scenes=df.Location.unique().tolist()
#df1.Location.unique().value_counts()

c=Counter(df.Location)
#print(c.most_common(100))
#len(c)
histo=c.most_common(5)
freq={}
for item,count in histo:
    freq[item]=count
freq
explode = (0.05,0.05,0.05,0.05,0.05)

plt.pie([float(v) for v in freq.values()], labels=[(k) for k in freq],
           autopct='%.2f',startangle=90, pctdistance=0.85, explode = explode,radius=3,textprops={'fontsize': 25})


centre_circle = plt.Circle((0,0),0.90,fc='white')
fig = plt.gcf()

fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
#ax1.axis('equal')  
plt.tight_layout()
#plt.title("Top 5 Shoot Locations on F.R.I.E.N.D.S.")

plt.show()






# from collections import Counter
# from matplotlib import pyplot as plt
# #print(scene_text)
# scenes=df.Location.unique().tolist()
# #df1.Location.unique().value_counts()

# c=Counter(df.Location)
# #print(c.most_common(100))
# #len(c)
# histo=c.most_common(5)
# freq={}
# for item,count in histo:
#     freq[item]=count
# freq
# explode = (0.05,0.05,0.05,0.05,0.05)
# # plt.rcParams["figure.figsize"] = (25,10)
# # plt.rcParams.update({'font.size': 30})

# plt.pie([float(v) for v in freq.values()], labels=[(k) for k in freq],
#            autopct='%.2f',startangle=90, pctdistance=0.85, explode = explode,radius=3,textprops={'fontsize': 35})


# centre_circle = plt.Circle((0,0),0.90,fc='white')
# fig = plt.gcf()

# fig.gca().add_artist(centre_circle)
# # Equal aspect ratio ensures that pie is drawn as a circle
# #ax1.axis('equal')  
# plt.tight_layout()
# #plt.title("Top 5 Shoot Locations on F.R.I.E.N.D.S.")

# plt.show()
```
